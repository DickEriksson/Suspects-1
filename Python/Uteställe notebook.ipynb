```python
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner

import folium # map rendering library
```

## 1. Loading the data into a DataFrame <a class="anchor" id="item1"></a>

Vi startar med att Load:a in vårt väderdata in i en DataFrame.

```python
# file path
file = 'väder.csv'

# New dataframe
df = pd.read_csv(file)

# Show new dataframe
df
```

## 2. Plot the data onto a map <a class="anchor" id="item2"></a>

Vi plot:ar vår data på karta och visualiserar "Väderpoäng" genom storleken på ringen.

```python
# Sverige
latitude = 63
longitude = 16

# Sverigekarta
sverigekarta = folium.Map(location=[latitude, longitude], zoom_start=5, width=750, height=700)

# add markers to map
for lat, lng, stad, score in zip(df['Latitud'], df['Longitud'], df['Stad'], df['Väderpoäng']):
    label = '{}'.format(stad)
    label = folium.Popup(label, parse_html=True)
    score = ((score + 10)**3)/150
    folium.CircleMarker(
        [lat, lng],
        radius=score,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(sverigekarta)  
    
sverigekarta
```

#### Sparar sverigekartan

```python
sverigekarta.save("sverigekarta.html")
```
